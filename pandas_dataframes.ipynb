### Filtering a Dataframe

```python
ipl.head()
```

```python
# find all the winners
pd.DataFrame(ipl[ipl["MatchNumber"]=="Final"][["WinningTeam","Season"]])
```

```python
# how many superovers finishes have occured
ipl[ipl["SuperOver"]=="Y"].shape[0] # 14 rows
```

```python
pd.DataFrame(ipl["WinningTeam"].values).drop_duplicates()
```

```python
# how many mathces has csk won in kolkata
ipl[(ipl["City"]=="Kolkata")&(ipl["WinningTeam"]=="Chennai Super Kings")].shape[0]
```

```python
# toss winner is match winner in percentage
ipl[(ipl["TossWinner"])==(ipl["WinningTeam"])].shape[0]*100/ipl.shape[0]
```

```python
# movies with rating higher than 8 and votes > 10000
movies[(movies["imdb_rating"]>8)&(movies["imdb_votes"]>10000)].shape[0]
```

```python
# action movies rating greater than 7.5
movies[(movies["genres"].str.split("|")).apply(lambda x : "Action" in x)&(movies["imdb_rating"]>7.5)].shape[0]
```
