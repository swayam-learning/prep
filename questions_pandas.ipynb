```python
import pandas as pd
import numpy as np
```

```python
ipl = pd.read_csv("deliveries.csv")
ipl.head()
```

```python
# find the top 10 batsman in terms of runs
batsman = ipl.groupby("batsman")
```

```python
batsman["batsman_runs"].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with maximum no of sixes
six_runs = ipl[ipl["batsman_runs"]==6]
six_runs.groupby("batsman")["batsman"].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find batsman with most number of 4's and 6's in last 5 over
overs = ipl[ipl["over"]>5]
runs = overs[(overs["batsman_runs"]==4)|(overs["batsman_runs"]==6)]
bats = runs.groupby("batsman")["batsman"].value_counts().sort_values(ascending=False)
bats.head(1)
```

```python
# find record of virat kohli against all the team
temp = ipl[ipl["batsman"]=="V Kohli"]
temp.groupby("bowling_team")["batsman_runs"].sum().reset_index()
```

```python
# create a function that return the highest score of any batsman
```

```python
ipl.head(1)
```

```python
def high_score(batsman):
    temp =  ipl[ipl["batsman"]==batsman]
    return temp.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1).values[0]
```

```python
high_score("MS Dhoni")
```
